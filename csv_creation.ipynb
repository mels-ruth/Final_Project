```python
import pandas as pd
```

```python
df = pd.read_csv("US_Accidents_March23.csv", 
                 usecols=["Severity", "Weather_Condition"]).dropna()

# Map weather conditions
df["Weather_Group"] = df["Weather_Condition"].map(lambda x: weather_map.get(x, "Other"))
agg = df.groupby(["Weather_Group", "Severity"]).size().reset_index(name="Count")

# Calculate proportion
agg['Proportion'] = agg.groupby('Weather_Group')['Count'].transform(lambda x: x / x.sum())

# Save
agg.to_csv("weather_severity_counts.csv", index=False)

print("CSV created")
```
